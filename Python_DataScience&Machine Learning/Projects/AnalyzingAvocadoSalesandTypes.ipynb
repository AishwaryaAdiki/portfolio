```python
TotalUsDrop = AvocadoData['region'] == "TotalUS"
NewAvocadoData = AvocadoData[~TotalUsDrop]
rowsDropped = ((AvocadoData.size) - (NewAvocadoData.size))/14
print("Number of rows dropped = " + str(rowsDropped))
```

```python
NewAvocadoData = NewAvocadoData.drop(['4046', '4225', '4770'], axis=1)
NewAvocadoData
```

## **Used dictionary comprehension (similar to list comprehension, but for dictionaries) to create a dictionary where the keys are the column names with a space between the words and values as the column names without the space. Used this dictionary to rename the columns of the dataframe.**

```python
new_names = {
    'Date': 'date',
    'AveragePrice': 'avg_price',
    'Total Volume': 'total_volume',
    'Total Bags': 'total_bags',
    'Small Bags': 'small_bags',
    'Large Bags': 'large_bags',
    'XLarge Bags': 'xlarge_bags',
    'type': 'avocado_type',
    'year': 'year',
    'EstimatedRevenue' : 'estimated_revenue',
    'region': 'region'
}

# Rename columns using the dictionary
NewAvocadoData = NewAvocadoData.rename(columns=new_names)
NewAvocadoData
```

## **Grouped, summed, and unstacked the Data:**

```python
# Group, sum, and unstack the data
NewAvocadoData_grouped = NewAvocadoData.groupby(['avocado_type','region','year']).sum()[['total_bags']].unstack(level=0)
NewAvocadoData_grouped
```

## **Convert the Date column to the datetime type. Filter the data so it contains just the rows for 2015, for the conventional type, and for the region named Albany. Store just the Date, Total Bags, and Small Bags columns in a new DataFrame:**

```python
# Convert the Date column to datetime type
NewAvocadoData['date'] = pd.to_datetime(NewAvocadoData['date'])

NewAvocadoData_filtered = NewAvocadoData[(NewAvocadoData['date'].dt.year == 2015) & (NewAvocadoData['avocado_type'] == 'conventional') & (NewAvocadoData['region'] == 'Albany')][['date', 'total_bags', 'small_bags']]
NewAvocadoData_filtered
```

## **Melt the data in the Total Bags and Small Bags columns, but not the values in the Date column. Name the column that contains the type of bag Bags, and name the column that contains the number of bags Count. Then, display the resulting DataFrame.**

```python
# Melt Total Bags and Small Bags columns
melted_df = pd.melt(NewAvocadoData, id_vars=['date'], value_vars=['total_bags', 'small_bags'], var_name='Bags', value_name='Count')
melted_df
```

## **Plot the melted data with Seaborn in a line plot, using the hue parameter to distinguish between the bag types**

```python
import seaborn as sns
sns.lineplot(x='date', y='Count', hue='Bags', data=melted_df)
```

## **Bin the data in the Total Volume column into four quantiles labeled â€˜poorâ€™, â€˜modestâ€™, â€˜goodâ€™, and â€˜excellentâ€™, and store the bin labels in a new column:**

```python
labels = ['poor', 'modest', 'good', 'excellent']
NewAvocadoData['volume_category'] = pd.qcut(NewAvocadoData['total_volume'], q=4, labels=labels)
NewAvocadoData
```

## **Plot the binned data by year using a Seaborn count plot**

```python
sns.countplot(x='year', hue='volume_category', data=NewAvocadoData)
```

## **Of the three bag sizes, which size sells the most? Hint: Melt the three bag size columns and plot the bags by year**

```python
melted_bags_df = pd.melt(NewAvocadoData, id_vars=['date', 'year'], value_vars=['small_bags', 'large_bags', 'xlarge_bags'], var_name='Bag Size', value_name='Count')

# Plot the bags by year using a Seaborn line plot
sns.lineplot(x='year', y='Count', hue='Bag Size', data=melted_bags_df)
```

## **Type of avocado that sells the most: **

```python
type_volume_df = NewAvocadoData.groupby('avocado_type')['total_volume'].sum().reset_index()

# Sort the resulting DataFrame by Total Volume in descending order
sorted_type_volume_df = type_volume_df.sort_values('total_volume', ascending=False)

# Print the type that sells the most
print(f"The type of avocado that sells the most is {sorted_type_volume_df.iloc[0]['avocado_type']}.")
```
