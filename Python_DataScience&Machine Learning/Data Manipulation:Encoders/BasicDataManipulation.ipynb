### **Restaurants in the order of highest protein per serving to lowest**

```python
sortedarray = nutr_df.sort_values(by=['Protein (g)'], ascending=False)
subarray = sortedarray[['Protein (g)','Fast Food Restaurant', 'Type']]
answer = subarray[subarray['Type'] == 'Burger'].groupby('Fast Food Restaurant')
answer.head()
```
